```python
def explore_data(dataset, start, end, header=False):
    dataset_slice = dataset[start:end]
    dataset_slice = dataset[start:end]    
    for row in dataset_slice:
        print(row)
        print('\n') ## adds a new (empty) line after each row

    if header is False:
        print('Number of rows:', len(dataset))
        print('Number of columns:', len(dataset[0]))
    else:
        dataset=dataset[1:]
        print('Number of rows:', len(dataset))
        print('Number of columns:', len(dataset[0]))
    return '\n'
```

## Part 2: Finding Most Popular App Genre, by User Rating

Guidelines:

"One way to find out what genres are the most popular (have the most users) is to calculate the average number of installs for each app genre. For the Google Play data set, we can find this information in the Installs column, but this information is missing for the App Store data set. As a workaround, we'll take the total number of user ratings as a proxy, which we can find in the rating_count_tot app.

Let's start with calculating the average number of user ratings per app genre on the App Store. To do that, we'll need to:

    -Isolate the apps of each genre.
    -Sum up the user ratings for the apps of that genre.
    -Divide the sum by the number of apps belonging to that genre 
    (not by the total number of apps)."


```python
def genre_table(dataset, genre_ind, user_rat_ind): 
    genre_cnt={}
    rating_total={}
    Rating_Mean={}
    for row in dataset:
        genre=row[genre_ind]
        rating=row[user_rat_ind]
        if rating=='NaN':
            rating_fl==0.0
        else:
            rating_fl=float(rating)
        #rating_cnt=float(row[rate_cnt_ind])
        if genre in genre_cnt:
            genre_cnt[genre]+=1
            rating_total[genre]+=rating_fl
        else:
            genre_cnt[genre]=1
            rating_total[genre]=rating_fl
    Rating_tuple=[]
    for key in genre_cnt:
        Rating_Mean[key]=rating_total[key]/genre_cnt[key]
        Rating_tuple.append([Rating_Mean[key], key])
    sorted_Rating_tuple=sorted(Rating_tuple, reverse = True)
    for i in range(len(sorted_Rating_tuple)-1):
        print(sorted_Rating_tuple[i])
    return ""
```

```python
print("Android App Ratings, Genre:\n")
print(genre_table(free_apps_data, genre_ind=9, user_rat_ind=2))
print("Android App Ratings, Category:\n")
print(genre_table(free_apps_data, genre_ind=1, user_rat_ind=2))
print("iOS App Ratings, Genre:\n")
print(genre_table(ifree_apps_data, genre_ind=11, user_rat_ind=7))
```

## Part 3: Finding Most Popular App Genre, by Installation Number

Guidelines:

"We have data about the number of installs for the Google Play market, so we should be able to get a clearer picture about genre popularity.

To perform computations, however, we'll need to convert each install number from string to float. This means we need to remove the commas and the plus characters, otherwise the conversion will fail and raise an error."

The Android data has installation number data, however the iOS data only has the number of reviews: we will use this as an approximation of install number.


```python
#Android Installation Number Data
Category=freq_table(free_apps_data, 1) #Category data
Install_Ave={}
Install_tuple=[]
for key in Category:
    total=0
    len_category=0
    for row in free_apps_data:
        category_app=row[1]
        if category_app==key:
            installs=row[5]
            installs=installs.replace('+',"")
            installs=installs.replace(',',"")
            installs=float(installs)
            total+=installs
            len_category+=1
    Ave=total/len_category
    Install_Ave[key]=Ave
    Install_tuple.append([Install_Ave[key], key])
sorted_Install_tuple=sorted(Install_tuple, reverse = True)
print("\nAverage Number of Android App Installations, by Category:\n")
for element in sorted_Install_tuple:
    print(element)
#Repeat for iOS Review Number Data
iCategory=freq_table(ifree_apps_data, 11) #Genre data
iInstall_Ave={}
iInstall_tuple=[]
for key in iCategory:
    itotal=0
    ilen_category=0
    for row in ifree_apps_data:
        icategory_app=row[11]
        if icategory_app==key:
            iinstalls=row[5]
            iinstalls=iinstalls.replace('+',"")
            iinstalls=iinstalls.replace(',',"")
            iinstalls=float(iinstalls)
            itotal+=iinstalls
            ilen_category+=1
    iAve=itotal/ilen_category
    iInstall_Ave[key]=iAve
    iInstall_tuple.append([iInstall_Ave[key], key])
isorted_Install_tuple=sorted(iInstall_tuple, reverse = True)
print("\nAverage Number of iOS App Installations/Reviews, by Category:\n")
for element in isorted_Install_tuple:
    print(element)
```
